<h1>1. Estructura Excepciones</h1>

Las **Excepciones** se utiliza para **evitar** cortes durante la ejecucin del programa.

```python
try:
    <cdigo>
    <cdigo>
    ..........

except:
    <cdigo>
    <cdigo>
    ..........
```

El comando **try** funciona de la siguiente manera:

- Primero, se ejecuta la clusula **try** (bloque de cdigo entre las palabras clave **try** y **except**.

- Si no se produce ninguna excepcin, se omite la clusula **except** y **try** finaliza la ejecucin de la instruccin.

- Si se produce una excepcin durante la ejecucin de la clusula **try**, se salta el resto de la clusul. Luego,si su tipo coincide con la **excepcin** nombrada despus de **except**, se ejecuta la clusula **except** y luego la ejecucin contina despus de **try**.

- Si ocurre una excepcin que no coincide con la excepcin nombrada en la clusula **except**, se pasa al bloque despus de **try**; si no se encuentra ningn controlador, es una excepcin no controlada y la ejecucin se detiene con un mensaje de **error**.

<h1>2. Estructura Funciones</h1>

**Por qu utilizar Funciones?**

- La creacin de una nueva funcin le da la oportunidad de nombrar un grupo de sentencias, lo que hace que su programa sea ms **fcil de leer y depurar**.

- Las funciones pueden hacer que un **programa sea ms pequeo** eliminando el cdigo repetitivo. Ms tarde, si haces un cambio, solo tienes que hacerlo en un lugar.

- Dividir un programa largo en funciones le permite **depurar las partes de una en una** y luego ensamblarlas en un todo de trabajo.

- Las funciones bien diseadas son a menudo **tiles para muchos programas**. Una vez que escribes y depuras uno, puedes reutilizarlo.

```python
def nombre_de_funcion(<argumento,s>):
    <cdigo>
    <cdigo>
    .....................
```

<h2>Hacer un programa que solicite ingreso de CUI correctamente:</h2>

```python
#Mi primera funcin validar dni
def validar_cui():
    while True:
        try:
            cui = int(input('Ingresar CUI: '))
            if len(str(cui)) == 8:
                print('\n\tCUI VLIDO')
                break
            else:
                print('\tError: Longitud de CUI no Vlido ... Intente de nuevo')
                continue
                
        except ValueError:
            print('\tError!!! No es un CUI vlido... Intente de Nuevo')
            
    return cui


#Programa MAIN principal

if __name__ == '__main__':
    titulo = ('   BENEFICIO PARA EL COMEDOR UNIVERSITARIO   ')
    print(titulo.center(50,'*'))
    print('\n\n')
    
    val_cui = validar_cui()
    
    print('\n\n El CUI Ingresado es: ==>',val_cui)
```
